```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv('Wine.csv')
df
```

```python
df.isnull().sum()
```

```python
x = df.drop('Customer_Segment',axis=1)
y = df['Customer_Segment']
```

```python
sc = StandardScaler()
s_scaled = sc.fit_transform(x)
```

```python
pca = PCA()
x_pca = pca.fit_transform(s_scaled)
```

```python
plt.scatter(x_pca[y==1][:,0],x_pca[y==1][:,1],c='r',label="Red Wine")
plt.scatter(x_pca[y==2][:,0],x_pca[y==2][:,1],c='b',label="White Wine")
plt.xlabel('Red Wine')
plt.ylabel('White Wine')
plt.legend()
plt.show()
```
